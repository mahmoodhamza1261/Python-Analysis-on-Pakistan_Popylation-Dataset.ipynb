```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data1=pd.read_csv("/content/pakistan_population.csv")
data2=pd.read_csv("/content/pakistan_population_forecast.csv")
data3=pd.read_csv("/content/population_pakistan_cities.csv")
```

1. What has been the trend in Pakistan&#39;s population growth over the years?

```python
# Create a DataFrame
df = data1

# Sort by Year in ascending order
df.sort_values('Year', inplace=True)

# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Population'], marker='o', linestyle='-', color='b')
plt.title('Trend in Pakistan\'s Population Growth Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(df['Year'])  # Show all years on the x-axis
plt.tight_layout()

# Show the plot
plt.show()
```

2. How has the yearly percentage change in population evolved from 2020 to 2050?

```python
year= data2[(data2['Year'] >= 2020) & (data2['Year'] <= 2050)]
plt.figure(figsize=(10, 6))
plt.plot(year['Year'], year['Yearly % Change'], marker='o', linestyle='-')
plt.title('Yearly Percentage Change in Population')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.grid(True)
plt.tight_layout()
plt.show()
```

11. Which cities have seen the most significant population growth?

```python
df = data3
plt.figure(figsize=(10, 6))

# Filter the DataFrame
filtered_df = df[df['Population'] > 788915]

# Plot the data using plt.scatter or plt.bar if you want city names on the x-axis
plt.scatter(filtered_df['Population'], filtered_df['City Name'], color='b')

# For a line plot (if numeric data is needed for x-axis):
# plt.plot(filtered_df['Population'], filtered_df['City Name'], marker='o', linestyle='-', color='b')

plt.xlabel('Population')
plt.ylabel('City Name')
plt.title('Cities with Population Greater than 788,915')
plt.show()
```

12. How does the population distribution vary across major cities in Pakistan?

```python
# Sort the DataFrame by Population in descending order
df = df.sort_values(by='Population', ascending=False)

# Plot the data using a bar plot
plt.figure(figsize=(12, 8))
plt.barh(df['City Name'], df['Population'], color='teal')

# Add labels and title
plt.xlabel('Population')
plt.ylabel('City Name')
plt.title('Population Distribution Across Major Cities in Pakistan')

# Show the plot
plt.show()
```
